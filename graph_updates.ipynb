```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# figure 1 data

fig1 = pd.DataFrame({"Project Type" : ["Transmission & Distribution", "Treatment", "Storage", "Source", "Other"],
                    "Money" : [200.8, 75.10, 36.90, 6, 2.30]})
```

```python
# figure 1

# plot
plt.bar("Project Type", "Money", data=fig1, zorder=2)

# gridline
plt.grid(axis="y", zorder=0)

# labels and title
plt.xlabel("Project Type")
plt.ylabel("Billions of Dollars")
plt.title("January 2007 Infrastructure Needs of Water Systems")

plt.xticks(rotation=45);
```

I do not think the original plot was too bad, I just think had too much going on as a lot of the aspects felt unnessary. So, I kept the bar plot style, but took away the labels on top of each bar as they were not needed, and cleaned up the y-axis a bit to not be as detailed. I also only kept the horizontal grid lines as the vertical ones seemed excessive and had all of the bars be a solid color instead of having a shadowy look like the original. I also rotated the labels on the x-axis to make them a bit easier to read.

```python
# figure 3 data
fig3 = pd.DataFrame({"Year": [1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006], 
                    "United States": [3.7, 4.5, 4.2, 4.5, 3.7, 0.8, 1.6, 2.7, 4.2, 3.5, 2.9],
                    "Euro Area": [1.5, 2.6, 2.9, 2.8, 3.7, 1.8, 0.9, 0.7, 1.9, 1.0, 1.8]})

fig3
```

```python
# figure 3

# plot the lines
plt.plot("Year", "United States", data=fig3, color="r", label="United States")
plt.plot("Year", "Euro Area", data=fig3, color="b", label="Euro Area")

# add a legend
plt.legend()

# add gridlines
plt.grid(axis="y", zorder=0)
plt.grid(axis="x", zorder=0)

# labels and title
plt.xlabel("Year")
plt.ylabel("GDP Growth (%)")
plt.title("Transatlantic Gulf GDP Growth");
```

Something I did not like about the original was you had to read the values on the block in order to know what the growth was, and because of the way the blocks are spaced, the United States looks like it has higher values even when it does not. To fix that, I decided to go with a line plot to make it much clearer where each point is and where it stands with the other area and added gridlines for the x and y axis for better interpretation. I kept the same color scheme because I felt it made sense in the context. 

```python
# figure 4 data

fig4 = pd.read_csv("./data/linegraph_data.csv")

# convert data to datetime
fig4["date"] = pd.to_datetime(fig4["date"], format="%y-%b")
fig4['date'] = fig4['date'].dt.strftime('%b %Y')
fig4 = fig4.pivot(index="date", columns="medium", values="revenue_thousands")

fig4 = fig4.sort_index(ascending=False)
fig4
```

```python
# grid layout
n_lines = fig4.shape[1]
n_cols = 2
n_rows = 4  

# create the subplots and flatten axes array
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 4 * n_rows), sharex=False, sharey=True)
axes = axes.flatten()

# plot each line in its respective subplot
for i, column in enumerate(fig4.columns):
    axes[i].plot(fig4.index, fig4[column], marker="o")
    axes[i].set_title(column, fontsize=12)
    axes[i].set_xticklabels("")
    axes[i].grid(alpha=0.5)
    
# hide unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis("off")  
    
# keep y-axis ticks on left side only
for i, ax in enumerate(axes):
    if i % n_cols != 0: 
        ax.tick_params(axis='y', labelleft=False)
        
# set axes labels and title
axes[-2].set_xlabel("Date", fontsize=14) 
axes[-2].set_xticks(fig4.index)
axes[-2].set_xticklabels(fig4.index, rotation=45, ha='right', fontsize=10)
        
axes[-3].set_xlabel("Date", fontsize=12)
axes[-3].set_xticks(fig4.index)
axes[-3].set_xticklabels(fig4.index, rotation=45, ha='right', fontsize=10)

axes[0].set_ylabel("Revenue (thousands of $)") 
axes[2].set_ylabel("Revenue (thousands of $)")
axes[4].set_ylabel("Revenue (thousands of $)")
axes[6].set_ylabel("Revenue (thousands of $)")

fig.suptitle("Revenue by Medium from October 2014 - April 2015", fontsize=15)

# adjust layout for proper spacing
plt.tight_layout(rect=[0, 0, 1, 0.98], h_pad=0.25)

plt.show()
```

For this graph, I felt like the line plot format was correct to show the change in data overtime, it was just too busy all together on one plot. So, I decided to split the graphs into a subplot showcasing each medium, this way they can be interpreted seperately or together. I made sure to keep the same y-axis range for easy comparison between plots and I think it is still easy to see which options are more or less popular, but now the trends for each medium are more clear as well.

```python
# figure 5 data

fig5 = pd.DataFrame({"Network": ["Fox News", "CNN", "New York Times", "Washington Post"], "Jan 2020": 
                     [112565, 141008, 103670, 87504], "Jan 2024": [85024, 112754, 80862, 48624]})

fig5.set_index("Network", inplace=True)
fig5
```

```python
# figure 5

fig5.plot.bar(zorder=2)

# axes adjustments
plt.xticks(rotation=0)
plt.ylim(30000, 145000)
plt.grid(axis="y")

# titles and labels
plt.ylabel("Unique Visitors")
plt.title("Number of Unique Visitors in January 2020 and January 2024");
```

I decided to change those 4 graphs into one bar graph. Although the data is longitudinal, it only features 2 points over 4 years, so I thought a bar graph would more clearly display the drop in the amount of unique visitors without losing out on other data points. The original graph is also slightly misleading as it looks like there has been a steady decline since January 2020, while it is likely that there would be increases and decreases between those 4 years. I also decided to zoom the graph in to better compare the drop in unique visitors between networks, this way it is still clear that The Washington Post had a significantly bigger drop than the other networks without having to explicitly state that.
